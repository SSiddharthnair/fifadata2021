**Before we get into the exploring, the column "Height" has values in ft. We need it in cm and in float datatype for further exploring.**

```python
#Changing the height type to float
df['Height']=df['Height'].str.replace('\'','.')
df['Height']=df['Height'].astype('float')
```

```python
#Changing ft to cms
def ft_to_cm(data):
    return(data*30.4)
df['Height']=df['Height'].apply(ft_to_cm)
df['Height']
```

## 10. Plotting Different feautures against one another (Scatter, Boxplot, Lineplot and Countplot).

### Lineplot

It is a type of chart which displays information as a series of data points called 'markers' connected by straight line segments. Here the lines are plotted between Age of the players and the average overall of all the players in that age category.

```python
grouped = df.groupby(df['Age'])
a=grouped['Overall'].agg(np.mean)
sns.lineplot(a)
plt.show()
```

**From the graph, we can observe that players between the age of 27 to 32 have a higher average overall stats when compared to players from other age categories. (we are not considering the value of age 45 because of the presence of only one player).**

### Countplot

A countplot basically counts the categories and returns a
count of their occurrences.


```python
sns.countplot(x=df['Age'], data=df)
plt.show()
```

**The most no of players who play the sport are aged 21.**

### Boxplot

It shows the distribution of the quantitative data that
represents the comparisons between variables. boxplot shows
the quartiles of the dataset while the whiskers extend to
show the rest of the distribution i.e. the dots indicating
the presence of outliers.


```python
sns.boxplot(x=df['Age'], y=df['Potential'], data=df)
plt.show()
```

**The observed result is that higer potential are seen in players at the age category of 17-22.**
